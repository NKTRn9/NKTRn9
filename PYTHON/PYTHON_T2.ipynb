```python
# mimp wpp to check the validity of passwod primary condition for password given 0
# 1.minimum 8 character 
#2. at least one alphabet should be lowercase
#3. atleast alphabet should be of upper case 
#4.atleast one digit betwen 0 to 9atleast on charcter from underscore,@and $

password=input("enter password= ")
upper1=0
lower1=0
character1=0
digit=0
if(len(password)<8):
    print("invalid password")
else:
    for i in password:
        if i.isupper():
            upper1=1
        elif i.islower():
            lower1=1
        elif i.isdigit():
            digit1=1
        elif i=="_" or i=="@" or i=="$":
            character1=1
    if upper1 + lower1 + character1 +digit1 ==4:
        print("pass")
    else:
        print("invakid")
```

```python
#wpp to count nu. of special eliment from  given list
# an element isspecial if removel of that element make the list balace
# the list will be balanced if sum of even index element =sum of odd index element
l=[5,5,2,5,8]
esum,osum,count=0,0,0
for i in range (len(l)):
    l1=l.copy()
    l1.pop(i)
    esum=sum(l1[::2])
    osum=sum(l1[1::2])
    if esum==osum:
        count+=1
print(count)
```

```python
# wpp to convert integer number into english speling 
d={10:"ten",20:"twenty",30:"thirty",40:"forty",50:"fifty",60:"sixty",70:"seventy",80:"eighy",90:"ninty",1:"one",2:"two",3:"three",4:"four",5:"fiive",6:"six",7:"seven",8:"eight",9:"nine",
   11:"eleven",12:"twelve",13:"thieteen",14:"forteen",15:"fifhteen",16:"sixteen",17:"seventeen",18:"eighteen",19:"nineteen"}
def int_to_eng(num):
    if num<20:
        return d[num]
    if num<100:
        if num%10==0:
            return d[num]
        else:
            return d[num//10*10]+ " "+d[num%10]
    if num<1000:
        if num%100==0:
            return d[num//100]+" "+"hundread"
        else:
            return d[num//100]+" "+"hundread"+" "+int_to_eng(num%100)
num=int(input("enter num= "))
print(int_to_eng(num))
            
```
