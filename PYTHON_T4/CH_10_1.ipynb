```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Bar plot
names=["Keyan","Krish","Karmit","Joy"]
marks=[85,20,91,56]
plt.bar(names,marks,width=0.5,color=["r","m","k","y"],label="Marks")
plt.title("Student Data",fontweight="bold",color="g")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Horizontal Bar plot
names=["Keyan","Krish","Karmit","Joy"]
marks=[85,20,91,56]
plt.barh(names,marks,height=0.5,color=["r","m","k","y"],label="Marks")
plt.title("Student Data",fontweight="bold",color="g")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#wapp to fill colour between income and expense line
# if income>expenses then fill green colour else red
month = np.array([i for i in range(1,13)])
income = np.array([5000,9000,6000,6000,10000,7000,6000,4000,4000,5000,6000,4000])
expense = np.array([6000,6000,8000,3000,6000,9000,7000,8000,6000,6000,4000,8000])
plt.plot(month,income)
plt.plot(month,expense)
plt.fill_between(month,income,expense,where=(income>expense),color="green",label="positive",interpolate=True)
plt.fill_between(month,income,expense,where=(income<=expense),color="r",label="negative",interpolate=True)

plt.show()
```
