# 633.  Implement the following hierarchy . The Staff function has name and salary as its data members, the derived class 
Teaching has subject as its data member and the class NonTeaching has department as its data member. The derived 
class method overrides (extends) the methods of the base class.

```python
class staff:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def display(self):
        print("name=",self.name+" salary=",self.salary)
    
class teaching(staff):
    def __init__(self,name,salary,subject):
        super().__init__(name,salary)
        self.subject=subject
    def display(self):
        super().display()
        print("subject=",self.subject)
        
class nonTeaching(staff):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department=department
    def display(self):
        super().display()
        print("department=",self.department)
    
            
```

```python
a=teaching("s",50,"t")
a.display()
b=nonTeaching("devansh gandu",00,"zero")
b.display()
```

# abstract class

```python
from abc import ABC,abstractmethod
class shap(ABC):
    def __init__(self,name):
        self.name=name
    @abstractmethod
    def drow(self):
        pass
class circle(shap):
    def __init__(self,name):
        super().__init__(name)
    def drow(self):
        print("drow in a circle")
class triangle(shap):
    def __init__(self,name):
        super().__init__(name)
    def drow(self):
        print("drow in a triangle")        
        
    
```

```python
# s=shap()  TypeError: 
c=circle("dada")
c.drow()
t=triangle("jadu")
t.drow()
```

```python
from abc import ABC,abstractmethod
class Employee(ABC):
    def __init__(self):
        self.email_id=0
        self.name=""
        self.salary=0
    def to_get_empdetails(self):
        self.eid=int(input("enter id = "))
        self.name=input("enter name = ")
        self.salary=int(input("enter salary = "))
    def print_e_details(self):
        print("emp_id=",self.eid)
        print("e_name=",self.name)
        print("e_salary=",self.salary)
    def return_salary(self):
        return self.salary
    @abstractmethod
    def emp_id(self):
        pass
    
class Perks(Employee):
    def __init__(self):
        super().__init__()
        self.da=0
        self.hra=0
        self.pf=0
    def calculate_perk(self):
        self.da=self.salary*0.35
        self.hra=self.hra*0.17
        self.pf=self.pf*0.12
    def display_perk(self):
        print("da= ",self.da)
        print("hra= ",self.hra)
        print("pf= ",self.pf)
        print("total of perk=",self.da+self.hra-self.pf)
    def emp_id(self):
        return self.eid
    
class Netsalary(Perks):
    def __init__(self):
        self.net_salary=0
        super().__init__()
    def calculat_net_salary(self):
        self.net_salary=self.salary+self.da+self.hra-self.pf
    def displayall(self):
        super().print_e_details()
        super().display_perk()
        print("net salary= ",self.net_salary)
        
```
