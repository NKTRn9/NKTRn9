# set font property's for title and label:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,220,190,280,230,290,300,320,340,200])
font1={"family":"serif","color":"red","size":20}
font2={"family":"serif","color":"green","size":20}


plt.plot(x,y,marker='^',ms=20,color="r")
plt.xlabel("average pulse",fontdict=font1)
plt.ylabel("calorie burange",fontdict=font1)
plt.title("sport watch data",fontdict=font2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,220,190,280,230,290,300,320,340,200])



plt.plot(x,y,marker='^',ms=20,color="r")
plt.xlabel("average pulse",fontsize=20,fontweight="bold")
plt.ylabel("calorie burange",fontsize=20)
plt.title("sport watch data",fontsize=20)

```

# Line Style

```python
# aa har ma thoi ek mcq 100% puchase   aa badhu linestyle ma lakhay che
#solid line: "-"
#doted line= ":" 
#dashed line= "--"
#adshed dotline= "-."
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,220,190,280,230,290,300,320,340,200])



plt.plot(x,y,marker='^',ms=20,color="r",linestyle="--",linewidth=3)
plt.xlabel("average pulse",fontsize=20,fontweight="bold")
plt.ylabel("calorie burange",fontsize=20)
plt.title("sport watch data",fontsize=20)

```

# 698. A Program to create a line graph to show the profits of a company in various years. The data is as mentioned: x axis as
years and y axis as profits (in Millions). X=[2012,2013,2014,2015,2016,2017] and y = [9,10,10.5,8.8,10.9,9.75]. Plot a line
chart with x axis as "Years" and y axis as "Profits (in Millions)" and title of the line chart as "XYZ Company". Also the
linestyle should be dashed one.


```python
x=[2012,2013,2014,2015,2016,2017]
y=[9,10,10.5,8.8,10.9,9.75]
plt.plot(x,y,linestyle="--")
plt.title("xyz company")
plt.xlabel("years")
plt.ylabel("profit in million")
plt.show()
```
