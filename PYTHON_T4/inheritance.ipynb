# 2.Operator Overloading

```python
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):    #print function no use kari tyare aa method kam aavase    
        return "x="+str(self.x)+"y="+str(self.y)
    def __add__(self,other):
        x=self.x+other.x
        y=self.y+other.y
        return point(x,y)
    def __sub__(self,other):
        x=self.x-other.x
        y=self.y-other.y
        return point(x,y)
    def __mul__(self,other):
        x=self.x*other.x
        y=self.y*other.y
        return point(x,y)
    def __truediv__(self,other):
        x=self.x/other.x
        y=self.y/other.y
        return point(x,y)
    def __floordiv__(self,other):
        x=self.x//other.x
        y=self.y//other.y
        return point(x,y)
    def __pow__(self,other):
        x=self.x**other.x
        y=self.y**other.y
        return point(x,y)
    def __it__(self,other):
        if self.x<other.x:
            if self.y<other.y:
                return True
        return False  
    def __gt__(self,other):
        if self.x>other.x:
            if self.y>other.y:
                return True
        return False
    def __le__(self,other):
        if self.x<=other.x:
            if self.y<=other.y:
                return True
        return False
    def __ge__(self,other):
        if self.x>=other.x:
            if self.y>=other.y:
                return True
        return False
    def __ne__(self,other):
        if self.x!=other.x:
            if self.y!=other.y:
                return True
        return False
    def __eq__(self,other):
        if self.x==other.x:
            if self.y==other.y:
                return True
        return False
    
```

```python
p1=point(6,31)
print(p1)
p2=point(3,4)
print(p2)
p3=p1+p2
print(p3)
print(p1-p2)
print(p1*p2)
print(p1/p2)
print(p1//p2)
print(p1**p2)
print("have sacho khel thase saru")
print(p1<p2)
print(p1>p2)
print(p1<=p2)
print(p1>=p2)
print(p1!=p2)
print(p1==p2)
```

# 3.Multi level inheritance:

```python
class dada:
    def one(self):
        print("helo this is dada")
class dadi:
    def two(self):
        print("helo this is dadi")
class papa(dada,dadi):
    def three(self):
        print("helo this is papa")
class mmi(dada,dadi):
    def four(self):
        print("helo this is mmi")
class bta(papa,mmi):
    def four(self):
        print("helo this is dikaro")        
```

```python
a=bta()
b=papa()
a.one()
b.two()
```

# 642.  Write a program to create a class called Data having “value” as its data member. Overload the (>) and the (<) operator for 
the class. Instantiate the class and compare the objects using _lt_ and _gt_.

```python
class Data:
    def __init__(self,value):
        self.value=value
    def __gt__(self,other):
        if self.value<other.value:
            return True
        return False
    def __lt__(self,other):
        if self.value>other.value:
            return True
        return False 
a=Data(5)
b=Data(6)
print(a>b)
print(a<b)
```

```python
class data:
    def __init__(self,value="None"):
        if value:
            self.value=value
            self.status=True
        else:
            self.status=False
    def __str__(self):
        return f"status ={self.status}"
        
```

```python
d1=data()
print(d1)
```
